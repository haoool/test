```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
data_wuhan = pd.read_excel(r'/home/haol/下载/武汉律师网分析.xlsx',sheet_name='武汉律师网')
```

```python
data_wuhan_lawfirms = pd.DataFrame(index = data_wuhan.place.unique())
```

```python
#律师人数 男女比例统计
for place in data_wuhan.place.unique():

    male_num = len(data_wuhan.loc[(data_wuhan.place==place) & (data_wuhan.sex=='男')])
    female_num = len(data_wuhan.loc[(data_wuhan.place==place) & (data_wuhan.sex=='女')])
    
    print(place, '女律师比例：{}'.format(np.round(female_num/(male_num+female_num), 2)))
    data_wuhan_lawfirms.loc[place, 'sex_ratio'] = np.round(female_num/(male_num+female_num), 4)
        
    data_wuhan_lawfirms.loc[place, 'total'] = male_num+female_num
```

```python
#学历
for place in data_wuhan.place.unique():
    total_num = len(data_wuhan.loc[(data_wuhan.place==place)])
    AboveMaster_num =0
    for edu in ['硕士', '法学硕士', '博士','研究生','法学博士']:
        AboveMaster_num += len(data_wuhan.loc[((data_wuhan.place==place) & (data_wuhan.education==edu))])

    print(place, '硕士以上学历占比：{}'.format(np.round(AboveMaster_num/total_num, 4)))
    data_wuhan_lawfirms.loc[place, 'education'] = np.round(AboveMaster_num/total_num, 4)
```

```python
#律师执业时长-median
for place in data_wuhan.place.unique():
    mean_year = np.mean(data_wuhan.loc[data_wuhan.place==place, '执业时长/年'])

    print(place, '律师平均执业时长：{}'.format(np.round(mean_year, 4)))
    data_wuhan_lawfirms.loc[place, 'experiment'] = np.round(mean_year, 4)
```

```python
from sklearn.cluster import DBSCAN, KMeans
import seaborn as sns
```

```python
%matplotlib inline
for feature in ['education', 'experiment','total']:
    X = np.array(data_wuhan_lawfirms[feature]).reshape(-1,1)
    kmeans = KMeans(n_clusters=10).fit(X)
    
    
    data_wuhan_lawfirms['label_'] = kmeans.labels_
    intra_labels = np.sort(data_wuhan_lawfirms.groupby('label_').mean()[feature])
    for label in data_wuhan_lawfirms.label_.unique():
        key_ = data_wuhan_lawfirms.loc[data_wuhan_lawfirms.label_==label, feature].mean()
        score = int(np.argwhere(intra_labels == key_)) + 1
        data_wuhan_lawfirms.loc[data_wuhan_lawfirms.label_==label, feature+'_score'] = score
        
    sns.catplot(y=feature,x=feature+'_score', data=data_wuhan_lawfirms, kind='strip')
    plt.xlabel('SCORE')
    plt.ylabel(feature)
    plt.show()
    plt.close('all')
```

```python
data_wuhan_lawfirms['SCORE'] = 0.6*data_wuhan_lawfirms.total_score + 0.25*data_wuhan_lawfirms.experiment_score + 0.15*data_wuhan_lawfirms.education_score
```
